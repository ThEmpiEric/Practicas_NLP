```python
# Extraer las fechas  
import numpy as np

path_limpias = r'C:\Users\ericl\anaconda3\envs\nlp\practicas\esteneograficas_limpias\*'
fecha_pattern = r'Presidencia de la República\s*\|\s*(\d{1,2}) de (\w+) de (\d{4})'


fechas = []

for file_pagina in glob.glob(path_limpias):
    with open(file_pagina, 'r', encoding='utf-8') as f:
        contenido = f.read()

        
        match = re.search(fecha_pattern, contenido)

        if match:
            dia, mes, anio = match.groups()
            fecha = f"{dia} de {mes} de {anio}"
            fechas.append(fecha)
        else:
            fechas.append(np.nan)  
```

### Metodo alternativo 

Esta parte experimental utiliza selenium, realiza el scraping de la pagina mas lento pero extrae los datos de las versiones esteneograficas directamente.    

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import time
import os

# Es necesario descargar el ChromeDriver necesario de tu navegador y copiar el path 
chromedriver_path = r"C:\Users\ericl\Downloads\chromedriver-win64\chromedriver-win64\chromedriver.exe"  
service = Service(executable_path=chromedriver_path)
driver = webdriver.Chrome(service=service)



# Carpeta para guardar las conferencias
if not os.path.exists("conferencias_txt"):
    os.makedirs("conferencias_txt")



def extraer_contenido_conferencia(url_conferencia, num_conferencia):
    '''Función para extraer el contenido de una conferencia y guardarla en un archivo .txt'''
    global driver
    try:
        driver.current_url  # Verifica si hay una ventana activa
    except:
        # Si no hay una ventana activa, reinicia el driver
        driver = webdriver.Chrome(service=service)
    
    driver.get(url_conferencia)
    time.sleep(5)
    
    try:
        # Buscamos el contenido de la conferencia en el div con la clase 'article-body'
        # En esta parte es previo hacer una inspeccion de la pagina a scrapear
        div_contenido = driver.find_element(By.CLASS_NAME, 'article-body')
        texto_conferencia = div_contenido.text
        nombre_archivo = f"conferencias_txt/conferencia_{num_conferencia}.txt"
        with open(nombre_archivo, "w", encoding="utf-8") as file:
            file.write(texto_conferencia)
        # print(f"Conferencia {num_conferencia} guardada en {nombre_archivo}")
    except Exception as e:
        print(f"Error al extraer la conferencia {num_conferencia}: {e}")



# URL de la página 1 de conferencias, solo consideramos una pagina 
url_pagina = "https://www.gob.mx/presidencia/es/archivo/articulos?idiom=es&order=DESC&page=1"
driver.get(url_pagina)
time.sleep(2)
# Inspeccionamos la parte en la que esta el contenido de las enteneograficas
enlaces_conferencias = driver.find_elements(By.CSS_SELECTOR, 'a[href*="version-estenografica"]')
urls_conferencias = [enlace.get_attribute('href') for enlace in enlaces_conferencias]



# Extrae el contenido de cada conferencia
num_conferencia = 1
for url_conferencia in urls_conferencias:
    # print(f"Procesando conferencia {num_conferencia}: {url_conferencia}")
    extraer_contenido_conferencia(url_conferencia, num_conferencia)
    num_conferencia += 1

driver.quit()





print("Proceso completo. Todas las conferencias han sido extraídas.")
```
